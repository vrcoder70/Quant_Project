# Mean Returns vs Log Returns

In trading and financial analysis, two common ways to measure the returns on investments are log returns (logarithmic returns) and simple returns (arithmetic returns). Each has its own advantages and is used in different contexts. Here’s a detailed comparison of the two:

### Simple Returns

Simple returns measure the percentage change in the value of an asset over a period. They are straightforward to calculate and interpret.

#### Formula:
$$ 
R_t = \frac{P_t - P_{t-1}}{P_{t-1}} 
$$
where:
- \( R_t \) is the simple return at time \( t \).
- \( P_t \) is the price at time \( t \).
- \( P_{t-1} \) is the price at time \( t-1 \).

### Log Returns

Log returns, or logarithmic returns, measure the continuously compounded rate of return. They are calculated using the natural logarithm of the ratio of consecutive prices.

#### Formula:
$$ r_t = \ln\left(\frac{P_t}{P_{t-1}}\right) $$
where:
- \( r_t \) is the log return at time \( t \).
- \( P_t \) is the price at time \( t \).
- \( P_{t-1} \) is the price at time \( t-1 \).


### Comparison

1. **Additivity**:
   - **Log Returns**: Log returns are additive over time. This means that the log return over a period is the sum of the log returns over sub-periods. This property is useful in time series analysis and for modeling returns over different time horizons.
     \[ r_{total} = r_1 + r_2 + \cdots + r_n \]
   - **Simple Returns**: Simple returns are not additive. The return over a period is not the simple sum of returns over sub-periods.

2. **Compounding**:
   - **Log Returns**: They assume continuous compounding, which is a more accurate representation of the return process over very small intervals.
   - **Simple Returns**: They assume discrete compounding, which is simpler but can be less accurate for high-frequency data.

3. **Symmetry**:
   - **Log Returns**: Log returns are symmetric, meaning that the return from \( P_{t-1} \) to \( P_t \) is the negative of the return from \( P_t \) to \( P_{t-1} \).
     \[ \ln\left(\frac{P_t}{P_{t-1}}\right) = -\ln\left(\frac{P_{t-1}}{P_t}\right) \]
   - **Simple Returns**: Simple returns do not have this property. A 50% drop in price is not offset by a subsequent 50% rise in price.

4. **Interpretation**:
   - **Simple Returns**: Easier to interpret for laypeople and are directly related to percentage changes in price.
   - **Log Returns**: More appropriate for mathematical modeling and statistical analysis, particularly when dealing with normal distributions and volatility.

5. **Statistical Properties**:
   - **Log Returns**: Often used in financial models because they tend to be normally distributed, which simplifies many types of statistical analysis.
   - **Simple Returns**: Can exhibit skewness and kurtosis, which complicates statistical modeling.

### Practical Example in Python

Let's calculate both simple and log returns for a series of stock prices:

```python
import pandas as pd
import numpy as np

# Sample data
data = {
    'date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04'],
    'AAPL': [150.00, 152.00, 153.00, 151.00],
    'AMD': [90.00, 92.00, 91.00, 93.00],
    'FOX': [40.00, 42.00, 41.00, 43.00]
}

df = pd.DataFrame(data)
df.set_index('date', inplace=True)

# Calculate simple returns
simple_returns = df.pct_change()

# Calculate log returns
log_returns = np.log(df / df.shift(1))

print("Simple Returns:")
print(simple_returns)

print("\nLog Returns:")
print(log_returns)
```

### Summary

- **Simple Returns** are intuitive and easy to interpret as percentage changes.
- **Log Returns** are more suitable for statistical modeling and provide additive properties over time.
- The choice between simple and log returns depends on the context and the type of analysis you are conducting. For basic performance tracking, simple returns are often sufficient. For more advanced financial modeling and time series analysis, log returns are preferred.

### Exponential Moving Average (EMA)

The Exponential Moving Average (EMA) is a type of moving average that places a greater weight and significance on the most recent data points. The weighting for each older data point decreases exponentially, hence the name.

#### Formula:

The formula for calculating the EMA at time \( t \) is:

\[ \text{EMA}_t = \alpha \cdot P_t + (1 - \alpha) \cdot \text{EMA}_{t-1} \]

where:
- \( \alpha \) is the smoothing factor, calculated as \( \alpha = \frac{2}{N+1} \) with \( N \) being the number of periods.
- \( P_t \) is the price (or data value) at time \( t \).
- \( \text{EMA}_{t-1} \) is the EMA value of the previous period.

The EMA reacts more quickly to recent price changes than the Simple Moving Average (SMA), making it a popular choice for trading strategies.

#### Calculation Example in Python:

```python
import pandas as pd

# Sample data
data = {'date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04'],
        'price': [100, 102, 101, 105]}
df = pd.DataFrame(data)
df.set_index('date', inplace=True)

# Calculate EMA with a span of 2 periods
df['EMA'] = df['price'].ewm(span=2, adjust=False).mean()

print(df)
```

### Rolling Method (Rolling Window)

The rolling method in time series analysis allows you to calculate a statistic (such as mean, sum, standard deviation) on a rolling window of fixed size.

#### Example Use Cases:
- **Rolling Mean**: Smoothens the data to identify trends.
- **Rolling Standard Deviation**: Measures volatility over a moving window.

#### Syntax in Pandas:
```python
rolling_window = df['price'].rolling(window=N)
```

where \( N \) is the number of periods in the rolling window.

#### Calculation Example in Python:

```python
import pandas as pd

# Sample data
data = {'date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04'],
        'price': [100, 102, 101, 105]}
df = pd.DataFrame(data)
df.set_index('date', inplace=True)

# Calculate rolling mean with a window of 2 periods
df['Rolling_Mean'] = df['price'].rolling(window=2).mean()

print(df)
```

### Detailed Comparison and Use Cases:

1. **EMA vs. Rolling Mean**:
   - **EMA**: Gives more weight to recent data points, making it more responsive to recent price changes.
   - **Rolling Mean**: Treats all data points within the window equally, making it less sensitive to recent changes but smoother.

2. **Use Cases**:
   - **EMA**: Preferred in trading strategies where recent price changes are more relevant (e.g., short-term trend following).
   - **Rolling Mean**: Used for identifying long-term trends, smoothing data, or calculating other rolling statistics (e.g., rolling volatility).

3. **Parameter Sensitivity**:
   - **EMA**: The sensitivity to recent changes can be adjusted via the span or smoothing factor.
   - **Rolling Mean**: The window size determines the level of smoothing.

### Summary:

- **EMA (Exponential Moving Average)**:
  - More weight on recent data.
  - Faster response to price changes.
  - Common in trading strategies.

- **Rolling Method**:
  - Equal weight to all data within the window.
  - Useful for calculating various rolling statistics.
  - Common in identifying trends and volatility over time.

# Both methods are essential tools in time series analysis, and the choice between them depends on the specific requirements of your analysis or trading strategy.

### Long Position and Short Position in Stock Trading

In stock trading, the terms "long position" and "short position" refer to different strategies investors use to potentially profit from the price movements of stocks.

### Long Position

A long position involves buying a stock with the expectation that its price will rise. The investor aims to sell the stock at a higher price in the future, thereby making a profit.

#### Key Points:

- **Buying**: The investor purchases shares of a stock.
- **Bullish Outlook**: The investor believes the stock price will increase.
- **Profit Mechanism**: Profit is realized when the stock is sold at a price higher than the purchase price.

#### Example:

1. **Initial Purchase**:
   - Investor buys 100 shares of XYZ Corp at $50 per share.
   - Total investment: \( 100 \times \$50 = \$5000 \).

2. **Price Increase**:
   - Stock price rises to $70 per share.

3. **Selling the Stock**:
   - Investor sells the 100 shares at $70 per share.
   - Total sale: \( 100 \times \$70 = \$7000 \).

4. **Profit Calculation**:
   - Profit: \( \$7000 - \$5000 = \$2000 \).

### Short Position

A short position involves selling a stock that the investor does not own, with the expectation that the stock price will decline. The investor aims to buy back the stock at a lower price in the future, thereby making a profit.

#### Key Points:

- **Selling**: The investor borrows shares of a stock and sells them.
- **Bearish Outlook**: The investor believes the stock price will decrease.
- **Profit Mechanism**: Profit is realized when the stock is repurchased at a price lower than the selling price.

#### Example:

1. **Initial Sale**:
   - Investor borrows 100 shares of XYZ Corp and sells them at $50 per share.
   - Total sale: \( 100 \times \$50 = \$5000 \).

2. **Price Decrease**:
   - Stock price falls to $30 per share.

3. **Buying Back the Stock**:
   - Investor buys back the 100 shares at $30 per share.
   - Total repurchase: \( 100 \times \$30 = \$3000 \).

4. **Returning the Borrowed Shares**:
   - Investor returns the 100 borrowed shares to the lender.

5. **Profit Calculation**:
   - Profit: \( \$5000 - \$3000 = \$2000 \).

### Risks:

- **Long Position**: The maximum loss is limited to the amount invested if the stock price drops to zero.
- **Short Position**: The maximum loss is theoretically unlimited since the stock price can rise indefinitely.

### Summary:

- **Long Position**: Buy low, sell high. Investor profits from an increase in the stock price.
- **Short Position**: Sell high, buy low. Investor profits from a decrease in the stock price.

### Practical Example in Trading:

1. **Long Position Example**:
   - **Initial Purchase**: Buy 100 shares of ABC Corp at $10.
   - **Price Rise**: Stock price rises to $15.
   - **Sell Shares**: Sell 100 shares at $15.
   - **Profit**: \( (100 \times 15) - (100 \times 10) = \$1500 - \$1000 = \$500 \).

2. **Short Position Example**:
   - **Initial Sale**: Borrow and sell 100 shares of ABC Corp at $10.
   - **Price Drop**: Stock price drops to $5.
   - **Buy Back Shares**: Buy back 100 shares at $5.
   - **Profit**: \( (100 \times 10) - (100 \times 5) = \$1000 - \$500 = \$500 \).

Both strategies are fundamental to stock trading, and traders choose one based on their market outlook and investment goals.

````python
def test_strategy(stock, start, end, SMA):
    df = yf.download(tickers=stock, start=start, end=end)
    data = df.Close.to_frame()
    data['returns'] = np.log(data.Close.div(data.Close.shift(1)))

    data['sma_s'] = data.Close.rolling(window=SMA[0]).mean()
    data['sma_l'] = data.Close.rolling(window=SMA[1]).mean()
    data.dropna(inplace=True)
    data['position'] = np.where(data['sma_s'] > data['sma_l'], 1, -1)
    data['strategy'] = data['returns']*data['position'].shift(1)

    ret = np.exp(data['strategy'].sum())
    std = data['strategy'].std() * np.sqrt(252)

    return ret,std
````

````python
test_strategy("SPY", '2000-01-01', '2024-01-01', (50,200))
````

````python
class SMABackTester():
    def __init__(self, symbol, start, end, SMA_S, SMA_L):
        self.symbol = symbol
        self.start = start
        self.end = end
        self.SMA_L = SMA_L
        self.SMA_S = SMA_S
        self.results = None
        self.get_data()

    def get_data(self):
        df = yf.download(tickers=self.symbol, start=self.start, end=self.end)
        data = df.Close.to_frame()
        data['returns'] = np.log(data.Close.div(data.Close.shift(1)))
    
        data['sma_s'] = data.Close.rolling(window=self.SMA_S).mean()
        data['sma_l'] = data.Close.rolling(window=self.SMA_L).mean()
        data.dropna(inplace=True)

        self.data = data


    def test_results(self):
        self.data['position'] = np.where(self.data['sma_s'] > self.data['sma_l'], 1, -1)
        self.data['strategy'] = self.data['returns']*self.data['position'].shift(1)

        self.data.dropna(inplace=True)
        self.data['returnbh'] = self.data['returns'].cumsum().apply(np.exp)
        self.data['stratey_return'] = self.data['strategy'].cumsum().apply(np.exp)

        self.results = (round(self.data['stratey_return'].iloc[-1]), round(self.data['returnbh'].iloc[-1]))

        return self.results

    def plot_result(self):
        if not self.results:
            self.test_results()
        title = f'{self.symbol} | SMA {self.SMA_S} | SMA {self.SMA_L}'
        self.data[['returnbh', 'stratey_return']].plot(title =title, figsize=(12,10), fontsize=12)
        # return self.results
````

````python
tester = SMABackTester('SPY','2000-01-01', '2020-01-01', 50, 100)
````

````python
tester.test_results()
````

````python
tester.plot_result()
````
